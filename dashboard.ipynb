### Global available data

```python
# Getting the data from the top 10 expensive neighbourhoods
avg_value_ngbhood = (to_data[["average_house_value"]].groupby(to_data["neighbourhood"]).mean())
avg_value_ngbhood = avg_value_ngbhood.sort_values(by="average_house_value", ascending=False)
top_10_ngbhood = avg_value_ngbhood.head(10)
top_10_ngbhood
# Calculate the mean number of dwelling types units per year
dwelling_type = (to_data[["single_detached_house","apartment_five_storeys_plus","movable_dwelling",
                           "semi_detached_house","row_house","duplex",
                           "apartment_five_storeys_less","other_house"]].groupby(to_data.index).sum())
                                
dwelling_type.head()

# Calculate the average monthly shelter costs for owned and rented dwellings
avg_shelter_cost = round(to_data[["shelter_costs_owned","shelter_costs_rented"]].groupby(to_data.index).mean(),2)
                                
avg_shelter_cost.head()
```

### Panel Visualization Functions

```python
# Define Panel visualization functions
def neighbourhood_map():
    """Neighbourhood Map"""
   
    avg_value_ngbhood = (to_data[["average_house_value"]].groupby(to_data["neighbourhood"]).mean())
    avg_value_ngbhood = avg_value_ngbhood.reset_index()
    
    avg_value_ngbhood = avg_value_ngbhood.drop (columns = "neighbourhood")
    
    location_value_ngbhood = pd.concat([df_neighbourhood_locations,avg_value_ngbhood], axis ="columns", join="inner")
    px.set_mapbox_access_token(map_box_api)

    location_ngbhood_plot = px.scatter_mapbox(
    location_value_ngbhood,
    lat = "lat",
    lon = "lon",
    color = "average_house_value",
    color_continuous_scale = px.colors.cyclical.IceFire,
    size = "average_house_value",
    zoom = 7)
    
    return location_ngbhood_plot
    
#def create_bar_chart(data, title, xlabel, ylabel, color):
    """
    Create a barplot based in the data argument.
    """
    
barchart_2001 = dwelling_type.loc[2001]
barchart_2001_plot =barchart_2001.hvplot.bar(title = "Dwelling Types in Toronto in 2001", color="red", ylabel="Dwelling Type Units", rot=90).opts(yformatter="%.0f")

barchart_2006 = dwelling_type.loc[2006]
barchart_2006_plot = barchart_2006.hvplot.bar(title = "Dwelling Types in Toronto in 2006", color="blue", ylabel="Dwelling Type Units", rot=90).opts(yformatter="%.0f")

barchart_2011 = dwelling_type.loc[2011]
barchart_2011_plot = barchart_2011.hvplot.bar(title = "Dwelling Types in Toronto in 2011", color="orange", ylabel="Dwelling Type Units", rot=90).opts(yformatter="%.0f")

barchart_2016 = dwelling_type.loc[2016]
barchart_2016_plot = barchart_2016.hvplot.bar(title = "Dwelling Types in Toronto in 2016", color="purple", ylabel="Dwelling Type Units", rot=90).opts(yformatter="%.0f")

barchart_plot_A = barchart_2001_plot + barchart_2011_plot
barchart_plot_B = barchart_2006_plot + barchart_2016_plot
     
    
    
def monthly_shelter_owned():
    """
    Create a line chart based in the data argument.
    """
    
    avg_shelter_owned = avg_shelter_cost["shelter_costs_owned"]
    owned_fig = avg_shelter_owned.hvplot.line(title="Average Monthly Shelter Cost for Owned Dwellings in Toronto", ylabel="Average Monthly Shelter Owned Costs", color ="blue")

    return owned_fig


def monthly_shelter_rented():
    """
    Create a line chart based in the data argument.
    """
    
    avg_shelter_rented = avg_shelter_cost["shelter_costs_rented"]
    rented_fig = avg_shelter_rented.hvplot.line(title="Average Monthly Shelter Cost for Rented Dwellings in Toronto", ylabel="Average Monthly Shelter Rented Costs", color ="orange")

    return rented_fig

def average_house_value():
    """Average house values per year."""
    
    average_house_value = round(to_data[["average_house_value"]].groupby(to_data.index).mean(),2)
    average_house_fig = average_house_value.hvplot.line(title="Average House Value in Toronto", ylabel="Average House Value", color ="blue").opts(yformatter="%.0f")


    return average_house_fig

def average_value_by_neighbourhood():
    """Average house values by neighbourhood."""
    
    ngbhood_house_value = to_data.filter(["neighbourhood","average_house_value"], axis=1)
    ngbhood_house_value_plot = ngbhood_house_value.hvplot.line(title="Average House Value per year by each neighbourhood", groupby="neighbourhood", ylabel="Average House Value", height = 500).opts(yformatter="%.0f")
    
    return ngbhood_house_value_plot 

def number_dwelling_types():
    """Number of dwelling types per year"""
    
    dwelling_type_ngbhood = to_data.filter(["neighbourhood","single_detached_house","apartment_five_storeys_plus","movable_dwelling",
                           "semi_detached_house","row_house","duplex",
                           "apartment_five_storeys_less","other_house"], axis=1)
    dwelling_type_ngbhood_plot = dwelling_type_ngbhood.hvplot.bar(title="Number of dwelling types per year by each neighbourhood", groupby="neighbourhood", 
                                                                  ylabel="Number of dwelling types", xlabel ="Year", rot=90, height=500)
    
    return dwelling_type_ngbhood_plot




def average_house_value_snapshot():
    """Average house value for all Toronto's neighbourhoods per year."""
    ngbhood_house_value = to_data.filter(["neighbourhood","average_house_value"], axis=1)
    ngbhood_house_value_reset = ngbhood_house_value.reset_index()

    ngbhood_house_value_bar = px.bar(ngbhood_house_value_reset,
                                x="neighbourhood",
                                y="average_house_value",
                                color ="average_house_value",
                                facet_row="year",
                                width=800, height=900,
                                labels={"average_house_value":"Avg. House Value"}
                                )
    
    return ngbhood_house_value_bar



def top_most_expensive_neighbourhoods():
    """Top 10 most expensive neighbourhoods."""
    
    avg_value_ngbhood = (to_data[["average_house_value"]].groupby(to_data["neighbourhood"]).mean())
    avg_value_ngbhood = avg_value_ngbhood.sort_values(by="average_house_value", ascending=False)
    top_10_ngbhood = avg_value_ngbhood.head(10)
    top_10_ngbhood
    
    top_10_ngbhood_plot = top_10_ngbhood.hvplot.bar(title="Top 10 Most Expensive Neighbourhoods in Toronto",ylabel="Average House Value", rot=90, height = 800).opts(yformatter="%.0f")
    
    return top_10_ngbhood_plot

def sunburts_cost_analysis():
    """Sunburst chart to conduct a costs analysis of most expensive neighbourhoods in Toronto per year."""
    cost_living_ngbhood = to_data.filter(["neighbourhood","shelter_costs_owned","shelter_costs_rented"], axis =1)
    
    list_top_ngbhood = top_10_ngbhood.reset_index()
    list_top_ngbhood=list_top_ngbhood["neighbourhood"].tolist()
    
    top_living_ngbhood = cost_living_ngbhood.query("neighbourhood in @list_top_ngbhood")
    
    fig = px.sunburst(top_living_ngbhood, path=[top_living_ngbhood.index,"neighbourhood"], values="shelter_costs_owned",
                  color="shelter_costs_owned", hover_data=["shelter_costs_owned"],
                  color_continuous_scale='RdBu',
                  color_continuous_midpoint=np.average(top_living_ngbhood["shelter_costs_owned"], weights=top_living_ngbhood["shelter_costs_owned"]),
                  title = "Costs Analysis of Most Expensive Neighbourhood in Toronto per Year - Owned")
    return fig

```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# Create a Title for the Dashboard
dashboard_title ="#Real Estate Analysis from 2001 to 2016 in Toronto"

# Define a welcome text
dashboard_welcome ="The dashboard presents a visual analysis of historical house values, dwelling types per neighbourhood and dwelling costs in Toronto, Ontario from 2001 to 2016"

dashboard_page = pn.Column (dashboard_title, dashboard_welcome, neighbourhood_map())
# Create a tab layout for the dashboard
yearly_analysis = pn.Column (barchart_plot_A, barchart_plot_B)
shelter_home_value = pn.Column (monthly_shelter_owned(), monthly_shelter_rented(),average_house_value())


house_value_dwelling = pn.Column (average_value_by_neighbourhood(), number_dwelling_types())
neighbourhood = pn.Row (average_house_value_snapshot(), house_value_dwelling)

top_neighbour = pn.Column (top_most_expensive_neighbourhoods(),sunburts_cost_analysis())
# Create the main dashboard
dashboard_tabs = pn.Tabs(             
                ("Home", dashboard_page),
                ("Yearly Market Analysis", yearly_analysis),
                ("Shelter Costs vs Home Value", shelter_home_value),
                ("Neighbourhood Analysis", neighbourhood),
                ("Top Expensive Neighbourhood", top_neighbour)
                        )

                
```
